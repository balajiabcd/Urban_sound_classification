```python
import numpy as np
import pandas as pd
import IPython.display as ipd

import matplotlib.pyplot as plt 
import seaborn as sns
sns.set()

import librosa
```

```python
pt1 = "4911-3-0-0.wav"
pt2 = "15564-2-0-1.wav"

x, sr = librosa.load(pt2)
```

# Librosa

```python
# Increasing the length of the length of audio file to 4 sec
while True:
    if len(x) < (sr*4+1):
        x = np.concatenate((x,[0]*sr,x), axis=0)
    if len(x) >= (sr*4+1):
        x = x[:sr*4+1]
        break

# Some iportant data
print(type(x)," ", type(sr))
print(len(x), " " ,sr, f", and time = {round(len(x)/sr,2)}")

# Entire x plotting
time = np.array(range(len(x)))/sr
plt.figure(figsize=(15,5))
plt.plot(time, x)

# First 1500 values of x plotting
a,b = 0,1500
plt.figure(figsize=(15,5))
plt.plot(time[a:b],x[a:b])
plt.show()

# Audio Player
ipd.Audio(x, rate = sr)
```

# Fast Fourior Transform (fft)

```python
# Applying ffc to x
y = np.fft.fft(x)

# Entire x plotting
time = np.array(range(len(y)))/sr
plt.figure(figsize=(15,5))
plt.plot(time, np.abs(y))

# First 1500 values of x plotting
a,b = 500,1000
plt.figure(figsize=(15,5))
plt.plot(time[a:b],np.abs(y[a:b]))
plt.show()

# Audio Player
ipd.Audio(np.abs(y), rate = sr/1)
```
